```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
deals=pd.read_csv('../data/deals.csv',
    header=None,
    names=["timestamp","deal_type","order_type","stop","limit","price","stop_price","limit_price","loss_or_profit"],
    dtype={
        "timestamp":str
    }
)
```

```python
deals.sort_values(by="timestamp")
```

```python
price=pd.read_csv('../data/price.csv',
    header=None,
    names=["datetime","timestamp","MID_OPEN","BID","OFFER","CHANGE","CHANGE_PCT","HIGH","LOW"],
    dtype={
        "timestamp":str,
        "datetime":str
    }
)
```

```python
price.sort_values(by="timestamp")
```

```python
deals_with_price=pd.merge(deals, price,how="left",on="timestamp",sort=True)
deals_with_price
```

```python
deals_with_price.to_csv('s3://%s/deals_with_price.csv' % BUCKET)
```

```python
labels=deals_with_price['deal_type'].apply(lambda x:1 if x=='limit' else 0)
labels
```

```python
deals_with_price['_order_type']=deals_with_price['order_type'].apply(lambda x:1 if x=='buy' else 0)

deals_with_price
```

```python
all_data=deals_with_price[["_order_type","MID_OPEN","BID","OFFER","CHANGE","CHANGE_PCT","HIGH","LOW","limit","stop"]]
all_data_with_label=deals_with_price[["_order_type","MID_OPEN","BID","OFFER","CHANGE","CHANGE_PCT","HIGH","LOW","limit","stop","deal_type"]]
count=len(all_data)

import math
training=math.floor(0.8*count)
evaluation=math.floor(0.2*count)
```

```python
training_data=all_data[0:training]
training_labels=labels[0:training]
training_data
```

```python
evaluation_data=all_data_with_label[(len(all_data_with_label)-evaluation):]
evaluation_data
```

```python
training_data.to_csv('training_data.csv',header=None,index=False)
evaluation_data.to_csv('evaluation_data.csv',header=None,index=False)
```

```python
training_data_np=training_data.to_numpy().astype('float32')
training_labels_np=training_labels.to_numpy().astype('int')
```

```python
training_labels_np
```
